```python
# generate sinusoid...

import numpy as np
import matplotlib.pyplot as plt

F = 10
Fs = 400
L = 50
x = np.zeros(L)
for n in range(L):
    x[n] = np.sin(2*np.pi*F*n/Fs)
#print(x)
#plt.plot(x)
plt.stem(x)
```

```python
#unit impulse...

n = range(-10, 10)
d = np.zeros(len(n))
d[len(n)//2] = 1
#plt.plot(d)
plt.stem(n, d)
```

```python
#unit step...

n = range(-10, 11)
d = np.zeros(len(n))
for i in range(len(n)//2):
    d[len(n)//2+i] = 1
#plt.plot(d)
plt.stem(n, d)
```

```python
# ramp...

n = range(-10, 11)
d = np.zeros(len(n))
for i in range(len(n)//2):
    d[len(n)//2+i] = 1+i
#plt.plot(d)
plt.stem(n, d)
```

```python
# ramp...(sir)

r = np.zeros(40)
for n in range(40):
    r[n] = n
#plt.plot(d)
plt.stem(r)
```

```python
# exp...

ur = np.zeros(50)
a = -0.95
for n in range(50):
    ur[n] = a**n
#plt.plot(d)
plt.stem(ur)
```
