- Generate a sinusoid

```python
import numpy as np
import matplotlib.pyplot as plt
f=10
Fs=400

x=np.zeros(50)
for n in range(50):
    x[n] = np.sin(2*np.pi*f*n/Fs)
print(x)

plt.plot(x)
```

```python
plt.stem(x)
```

- Unit Impulse

```python
n=range(-10,11)
d=np.zeros(len(n))
d[len(n)//2]=1
plt.plot(n,d)
```

```python
plt.stem(n,d)
```

- Unit Step

```python
n=range(-10,11)
d=np.zeros(len(n))
for i in range(len(n)//2):
    d[len(n)//2+i]=1
plt.plot(n,d)
```

```python
plt.stem(n,d)
```

- Ramp

```python
n=range(-10,11)
r=np.zeros(40)
for i in range(40):
    r[i]=i
plt.plot(r)
```

```python
plt.stem(r)
```

- exponential

```python
ur = np.zeros(50)
a=0.8
for n in range(50):
    ur[n]=(a**n)
plt.stem(ur)
```
